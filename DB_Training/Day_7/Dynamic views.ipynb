```sql
CREATE OR REPLACE FUNCTION dev.naval_gold.datamask(x STRING)
  RETURNS STRING
  RETURN CONCAT(REPEAT("*", LENGTH(x) - 2), RIGHT(x, 2)
); 
```

```sql
select naval_gold.datamask('naval@gmail.com') as email
```
